## Материальная точка движется вдоль оси OX по закону: $ x(t) = 4 \cdot t $
<font size="3">
<br> <b> Чему равна скорость материальной точки? </b>
<br> <b>Какой путь она пройдёт за 7 с движения? </b>
   
**Постройте графики зависимости (линейные диаграммы):**
- скорости от времени
- координаты от времени
</font>

### Импорт модулей

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

<font size='3'>
<b> Материальная точка будет двигаться прямолинейно, следовательно, справедливы следующие формулы: </b>

Скорость материальной точки: $ v = \frac{S}{t} $

Путь, пройденный материальной точкой: $ \Delta S = v * t $
</font>

```python
v, x, t = [0], [0], [0]
time = 7
for i in range(1, 8):
    xt = 4*i
    v2 = xt/i
    v.append(v2)
    t.append(i)
    x.append(xt)
print('Скорость мат. точки равна: ', v2, 'м/c')
print('Путь, пройденный за 7 с движения: ', v2*time,'м')
```

### Построим линейные диаграммы по полученным данным:

```python
plt.grid()
plt.title('График зависимости координаты от времени')
plt.xlabel('t, с.')
plt.ylabel('x, м.')
plt.plot(t, x)
plt.scatter(t, x)
plt.show()

plt.grid()
plt.title('График зависимости скорости от времени')
plt.xlabel('t, с.')
plt.ylabel('v, м/c')
plt.plot(t, v)
plt.scatter(t, v)
plt.show()
```
